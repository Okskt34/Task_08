```python
import pandas as pd
```

```python
names = ['Аксьонов', 'Буханчук', 'Ілюк', 'Ковтун', 'Кантур', 'Климовець', 'Колешня', 'Коркушко', 'Кучерявий', 'Межінський', 'Нечай', 'Новик', 'Старовойт', 'Степанюк', 'Терещенко']
progress = {
      'gpa':     [70, 80, 75, 60, 65, 85, 80, 67, 70, 72, 83, 85, 84, 90, 67],
      'current_exam':  [80, 85, 82, 73, 76, 78, 77, 90, 91, 80, 75, 76, 65, 78, 90],
      'number_attempts': [1, 1, 1, 2, 2, 1, 1, 1, 1, 2, 1, 1, 1, 1, 1]
     }
progress_df = pd.DataFrame(data=progress, index=names)
progress_df
```

```python
progress_df['course'] = 3
progress_df
```

```python
min_gpa = progress_df['gpa'].min()
max_gpa = progress_df['gpa'].max()
mean_gpa = progress_df['gpa'].mean()
median_gpa = progress_df['gpa'].median()

print("Мінімальний середній бал студентів:", min_gpa)
print("Максимальний середній бал студентів:", max_gpa)
print("Середній середній бал студентів:", mean_gpa)
print("Медіанний середній бал студентів:", median_gpa)
```

```python
passed_first_attempt = progress_df[progress_df['number_attempts'] == 1]
count_passed_first_attempt = len(passed_first_attempt)

print("Кількість студентів, що здали поточний екзамен з першого разу:", count_passed_first_attempt)
```

```python
passed_first_attempt = progress_df[progress_df['number_attempts'] == 1]
count_passed_first_attempt = passed_first_attempt['number_attempts'].value_counts()

print(count_passed_first_attempt)
```
